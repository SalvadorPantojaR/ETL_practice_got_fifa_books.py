```python
import json
import pandas as pd
import os

files = [
    r"C:\Users\chavs\Downloads\GOT Subtitles\season1.json",
    r"C:\Users\chavs\Downloads\GOT Subtitles\season2.json",
    r"C:\Users\chavs\Downloads\GOT Subtitles\season3.json",
    r"C:\Users\chavs\Downloads\GOT Subtitles\season4.json",
    r"C:\Users\chavs\Downloads\GOT Subtitles\season5.json",
    r"C:\Users\chavs\Downloads\GOT Subtitles\season6.json",
    r"C:\Users\chavs\Downloads\GOT Subtitles\season7.json"
]

def load_json(file):
    """Carga un archivo JSON y lo convierte en una lista de diccionarios si el archivo existe."""
    if not os.path.exists(file):
        print(f"Error: No se encontró el archivo {file}")
        return []
    
    with open(file, 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    subtitles = []
    for episode, lines in data.items():
        for idx, text in lines.items():
            subtitles.append({"episode": episode, "line_number": int(idx), "text": text})
    
    return subtitles


df_list = []
for file in files:
    data = load_json(file)
    if data:  
        df = pd.DataFrame(data)
        df_list.append(df)

for i, df in enumerate(df_list):
    print(f"\nPrimeras 10 filas de {files[i]}:")
    print(df.head(10))
    print(f"\nÚltimas 10 filas de {files[i]}:")
    print(df.tail(10))

for i, df in enumerate(df_list):
    selected_columns = df[["episode", "line_number", "text"]]  # Seleccionar 3 columnas relevantes
    output_file = os.path.join(r"C:\Users\chavs\Downloads\GOT Subtitles", f"processed_{os.path.basename(files[i]).replace('.json', '.csv')}")
    selected_columns.to_csv(output_file, index=False)
    print(f"Archivo generado: {output_file}")
```
